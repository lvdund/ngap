```python
import shutil

def delete_all_in_folder(folder_path):
    # Kiểm tra xem thư mục có tồn tại không
    if os.path.exists(folder_path):
        # Xóa tất cả nội dung trong thư mục
        shutil.rmtree(folder_path)
        # Tạo lại thư mục rỗng nếu cần thiết
        os.makedirs(folder_path)
        print(f"All contents in the folder '{folder_path}' have been deleted.")
    else:
        print(f"The folder '{folder_path}' does not exist.")

def delete_all_except(filepath, folder_path):
    # Duyệt qua tất cả các tập tin và thư mục trong folder_path
    for filename in os.listdir(folder_path):
        file_path = os.path.join(folder_path, filename)
        
        # Kiểm tra nếu filename không phải là file cần giữ lại
        if filename != filepath:
            # Kiểm tra nếu là một file và xóa nó
            if os.path.isfile(file_path) or os.path.islink(file_path):
                os.unlink(file_path)
            # Kiểm tra nếu là một thư mục và xóa nó
            elif os.path.isdir(file_path):
                shutil.rmtree(file_path)

# Ví dụ sử dụng
folder_path = './cache'
delete_all_in_folder(folder_path)
folder_path = '../ie'
delete_all_in_folder(folder_path)
```

```python
def extract_number_and_inner_text(text):
    # Sử dụng regex để lấy số đầu tiên và chuỗi bên trong dấu < >
    match = re.search(r'(\d+)\.\.<(.*?)>', text)
    if match:
        return match.group(1), match.group(2)
    return None, None
```

```python
# check type : like "9.3.1.197"
def is_type_exist(s):
    pattern = r'^\d+\.\d+\.\d+\.\d+$'
    return re.match(pattern, s) is not None

def starts_with_number(s):
    # Kiểm tra xem chuỗi có bắt đầu với số không
    return re.match(r'^\d+', s) is not None
```

```python
def get_cells_7_cols(cells):
    iei = cells[0].text
    level = iei.count('>')
    iei = clean_string(iei)
    if starts_with_number(iei):
        iei = "Ie" + iei
    presence = cells[1].text
    range = cells[2].text
    typeCell = cells[3].text
    if typeCell.startswith("5G"):
        typeCell = "Ie" + typeCell
    if is_type_exist(cells[3].text) or typeCell.startswith("9."):
        typeCell = iei
    elif re.search(r'\b\d{1,2}\.\d{1,2}\.\d{1,2}\.\d{1,3}\b', typeCell):
        # "some string 9.3.1.197"
        typeCell = re.sub(r'\d+\.\d+\.\d+\.\d+', '', typeCell).strip()
        typeCell = clean_string(typeCell)
    else:
        typeCell = typeCell.replace('\n', '')
    return {
        "iei" : iei,
        "level": str(level),
        "presence" : presence,
        "range" : range,
        "type" : typeCell,
    }

def write_cells_to_file(file, name, cells):
    write_file(file, name + ".append({ \"iei\" : \"" + cells["iei"] + \
        "\", \"level\" : \"" + cells["level"] + \
        # "\", \"presence\" : \"" + cells["presence"] + \
        "\", \"range\" : \"" + cells["range"] + \
        "\", \"type\" : \"" + cells["type"] + "\"})\n")

def write_file(file, string):
    file.write(string)
```

```python
def gen_type(vartype, range):
    match1 = re.search(r'\(SIZE\((\d+)\)\)', vartype) # SIZE(123)
    match2 = re.search(r'SIZE\((\d+)\.\.(\d+)', vartype) # SIZE(1..150
    match3 = re.search(r'INTEGER \((\d+)\.\.(\d+)', vartype) # INTEGER (1..100
    match4 = re.search(r'ENUMERATED \(([^)]+)\)', vartype)    # ENUMERATED (s1, s2, s3, s4, ...)
    type = ""
    min_size, max_size = 0, 0
    if match1:
        min_size = int(match1.group(1))
        max_size = min_size
        type = "[]byte"
    if match2:
        min_size = int(match2.group(1))
        max_size = int(match2.group(2))
        type = "[]byte"
    if match3:
        min_size = int(match3.group(1))
        max_size = int(match3.group(2))
        if "217" in vartype:
            max_size = 65535
            type = "Integer"
        elif "232" in vartype:
            max_size = 4294967295
            type = "Integer"
        elif "264" in vartype:
            max_size = 18446744073709551615
            type = "Integer"
        elif "240" in vartype:
            max_size = 2**40-1
            type = "Integer"
        elif max_size > 0 and max_size < 256:
            type = "Integer"
        elif max_size > 255 and max_size < 65536:
            type = "Integer"
        elif max_size > 65535 and max_size < 4294967296:
            type = "Integer"
        elif max_size > 4294967295 and max_size < 18446744073709551616:
            type = "Integer"
        elif "INTEGER (0.. maxNRARFCN)" in vartype:
            max_size = 3279165
            type = "uint32"
    if match4 or "ENUMERATED" in vartype or vartype == "NULL":
        type = "[]byte"
    if "BIT STRING" in vartype:
        type = "BitString"
    if "CHOICE (" in vartype or "VisibleString" in vartype:
        type = "[]byte"
    if vartype == "OCTET STRING" or "OCTET STRING" in vartype:
        type = "OctetString"
    if "INTEGER" in vartype:
        type = "Integer"
    if "Ie9.3.3.1" in vartype:
        type = "Tai"
    if "Ie9.2.13" in vartype:
        type = "MessageType"
    if "Ie9.3.3.21" in vartype:
        type = "AmfName"
    if "Ie9.3.1.74a" in vartype:
        type = "UeRadioCapabilityEUtraFormat"
    if vartype == "CpTransportLayerAddress":
        type = "CpTransportLayerInformation"
    if vartype == "QosFlowIndicator":
        type = "QosFlowIdentifier"
    if vartype == "ContainingTheUeContextSuspendRequestTransferIeSpecifiedInSubclause":
        type = "UeContextSuspendRequestTransfer"
    if type == "":
        type = vartype
    return type, min_size, max_size
```

```python
# write struct to go file
defined_struct = []
def string_struct_go(ies, name):
    if starts_with_number(name):
        name = "Ie" + name
    iesSubLv = {}
    header = "package ie\n"
    structGo = ""

    for i1 in range(len(ies)):
        ie1 = ies[i1]
        ieLv = []

        if i1 < len(ies) - 1:
            if int(ies[i1 + 1]["level"]) - int(ie1["level"]) <= 0 and ie1["type"] == "":
                ie1["type"] = ie1["iei"]
            elif int(ies[i1 + 1]["level"]) - int(ie1["level"]) == 1 and ie1["iei"].endswith("List") and ies[i1 + 1]["iei"].endswith("Item") and ie1["type"] == "" and ies[i1 + 1]["type"] == "":
                ie1["type"] = "[]" + ies[i1 + 1]["iei"]
                # ies[i1 + 1]["type"] = ies[i1 + 1]["iei"]
            elif int(ies[i1 + 1]["level"]) - int(ie1["level"]) == 1 and ie1["iei"].endswith("List") and ie1["range"] != "" and ies[i1 + 1]["range"] == "" and ie1["type"] == "[]":
                ie1["type"] = "[]" + ies[i1 + 1]["iei"]
        
        for ie2 in ies[i1+1:]:
            if int(ie2["level"]) - int(ie1["level"]) == 1:
                ieLv.append(ie2)
            elif int(ie2["level"]) - int(ie1["level"]) <= 0:
                break
        
        if len(ieLv) > 0 and ie1["type"] == "":
            ie1["type"] = ie1["iei"]
            iesSubLv[ie1["iei"]] = ieLv
    
    mainIeLv = []
    for i1 in range(len(ies)):
        if ies[i1]["level"] == "0":
            mainIeLv.append(ies[i1])
    if len(mainIeLv) > 0:
        # iesSubLv[name] = ieLv
        structGo += sub_struct(mainIeLv, name, True)

    if len(iesSubLv) > 0:
        for nameSub, iesub in iesSubLv.items():
            if len(iesub) > 0:
                structGo += sub_struct(iesub, nameSub, False)
    if "aper" in structGo:
        return header + "\nimport \"ngap/aper\"\n" + structGo
    return header + structGo

def sub_struct(ies, name, isMainStruct):
    structlv = "\ntype " + str(name) + " struct {\n"
    importAper = False
    if name in defined_struct and isMainStruct == False:
        return ""
    for i in range(len(ies)):
        ie = ies[i]
        type, min_size, max_size = gen_type(ie["type"], ie["range"])
        if max_size == 0:
            max_size = 150
        if type == name and len(ies) == 1:
            return ""
            # continue
        if type == "Integer":
            importAper = True
            structlv += ie["iei"] + "\taper." + type + "\t" + "//`Integer:\"valueLB:" + str(min_size) + ",valueUB:" + str(max_size) + "\"`\n"
        elif type == "OctetString":
            importAper = True
            structlv += ie["iei"] + "\taper." + type + "\t" + "//`octetstring:\"sizeLB:" + str(min_size) + ",sizeUB:" + str(max_size) + "\"`\n"
        elif type == "BitString":
            importAper = True
            structlv += ie["iei"] + "\taper." + type + "\t" + "//`bitstring:\"sizeLB:" + str(min_size) + ",sizeUB:" + str(max_size) + "\"`\n"
        else:
            structlv += ie["iei"] + "\t" + type + "\t" + "//`bitstring:\"sizeLB:" + str(min_size) + ",sizeUB:" + str(max_size) + "\"`\n"
        if i + 1 == len(ies):
            structlv += "}\n"
    defined_struct.append(name)
    if structlv == "\ntype " + str(name) + " struct {\n":
        return ""
    return structlv
```
